```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("jobmarket.csv")
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df,x='job_title')
plt.title("Most Common AI & Data Science Job Titles")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df,x='job_title',order=df['job_title'].value_counts().index)
plt.title("Distribution of AI & Data Science Job Roles",fontsize=16)
plt.xlabel("Job Title")
plt.ylabel("Number of Jobs")
plt.xticks(rotation=45)
plt.show()
```

```python
df['salary'].describe()
```

```python
avg_salary=df.groupby('job_title')['salary'].mean()
print(avg_salary)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df,x='job_title',y='salary')
plt.title("Average Salary by Job Ttile",fontsize=16)
plt.xlabel("Job Title")
plt.ylabel("Salary")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df,x='remote_type')
plt.title("Remote Work Distribution in AI & Data Science Jobs",fontsize=16)
plt.xlabel("Work Type")
plt.ylabel("Number of jobs")
plt.show()
```

```python
remote_counts=df['remote_type'].value_counts()
print(remote_counts)
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df,x='country')
plt.title("Country-wise AI & Data Science Job Distribution",fontsize=16)
plt.xlabel("Country")
plt.ylabel("Number of Jobs")
plt.show()
```

```python
country_counts=df['country'].value_counts()
print(country_counts)
```

```python
skills={'Python':df['skills_python'].sum(),'SQL':df['skills_sql'].sum(),'Machine Learning':df['skills_ml'].sum(),'Deep Learning':df['skills_deep_learning'].sum(),'Cloud':df['skills_cloud'].sum()}
print(skills)
```

```python
skills_df=pd.DataFrame(list(skills.items()),columns=['Skills','Count'])
plt.figure(figsize=(10,5))
sns.barplot(data=skills_df,x='Skills',y='Count')
plt.title("Most In-Demand Skills in AI & Data Science Jobs",fontsize=16)
plt.xlabel("Skills")
plt.ylabel("Demand Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df,x='experience_level')
plt.title("Experience Level Distribution",fontsize=16)
plt.xlabel("Experience Level")
plt.ylabel("Number of Jobs")
plt.show()
```

```python
experience_counts=df['experience_level'].value_counts()
print(experience_counts)
```

```python
salary_exp=df.groupby('experience_level')['salary'].mean()
print(salary_exp)
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df,x='experience_level',y='salary')
plt.title("Average Salary by Experience Level",fontsize=16)
plt.xlabel("Experience Level")
plt.ylabel("Average Salary")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df,x='education_level')
plt.title("Education Level Distribution",fontsize=16)
plt.xlabel("Education Level")
plt.ylabel("Number of Jobs")
plt.show()
```

```python
education_counts=df['education_level'].value_counts()
print(education_counts)
```

```python
plt.figure(figsize=(12,8))
correlation=df.corr()
sns.heatmap(correlation,annot=True)
plt.title("Correlation Heatmap of AI Job Dataset",fontsize=16)
plt.show()
```
